```python
def gensimmod(data, article_name):
    
    '''
    
    This function runs the LDA model and returns two output - 
    - output1 gives us the count of topics in which a particular document falls into. It has below variables:
        [cnt_of_topics_doc_appeared", "documents", "tokens"]
        
    - output2 gives us the percentage contribution along with the dominant topic in which a particular document falls in.
    That is to say, some documents are associated with multiple topics and have certain contribution in them. 
    The output here gives us the topic to why they are most associated with. 
    Column names in output2:-
    ['Doc_num', 'Dominant_Topic', 'Perc_Contribution', 'Topic_Keywords', 'orig_contents']
    
    '''

    
    'Extract Article content for topic modelling'
    reviews = data['Article_content']  # Extract Article content for topic modelling
    reviews = list(reviews)
    
    'Import WordNet Lemmatizer to lemmatize documents'
    lemma = WordNetLemmatizer()
    stop_words = stopwords.words('english')
    
    'User Named Entity Recognition from Spacy Module and get entities names, such as person, apartment, place etc'
    labels11 = ['Token', ' IOB', 'Entity']
    ner = pd.DataFrame()
    for i in range(0, len(reviews)):
        doc_1 = nlp(reviews[i]) # use nlp() from spacy module
        ner_1 = pd.DataFrame([(X, X.ent_iob_, X.ent_type_) for X in doc_1], columns = labels11)
        ner_1['doc_num'] = i
        ner = ner.append(ner_1) # Store entities in ner dataframe
        
    WordstoExclude = ner_1.loc[ner_1['Entity'] == 'ORG']  # ORG, organisation
    excludewords = WordstoExclude.Token
    excludewords = [str(i).lower() for i in excludewords]
    
    words = [] 
    for i in excludewords:
        if not i in words:
            words.append(i);
    
    
    'text cleaning of articles data'
    stop_words.extend(words) # Extending stop words list by adding Named Entity = ORG to it.
    stop_words.extend(['said', 'r', 'delivery', 'get', 'crore', 'company', 'rathi', 'say'])
    #Clean Text 
    def textClean(text0):
        text1 = [strip_punctuation(doc) for doc in text0]
        text1 = [strip_tags(doc) for doc in text1]
        text1 = [strip_numeric(doc) for doc in text1]
        text1 = [[" ".join([i for i in doc.lower().split() if i not in stop_words])] for doc in text1]
        #text2 = [[word for word in ' '.join(doc).split()] for doc in text1]
        normalized = [[" ".join([lemma.lemmatize(word) for word in ' '.join(doc).split()])] for doc in text1]
        return normalized
    
    corpus1 = textClean(reviews)  # Cleaned corpus .
    
    'word tokenization in corpus' 
    corpus11 = [[word for word in ' '.join(doc).split()] for doc in corpus1]  # word_tokenize 
    #corpus11[1]
    
    
    'Create Dictionary and Corpus required for LDA as inputs'
    id2word = corpora.Dictionary(corpus11)  # Create Dictionary
    corpus = [id2word.doc2bow(text) for text in corpus11]  # Building gensim corpus. TF DTM creation.
    #print(corpus[:1])  # View one doc in abstract form
    #id2word[2]
    
    
    '''
    A topic model is a sort of probabilistic generative model to find the representative topic of a document/corpus which 
    in our case , we are using latent Dirichlet allocation (LDA) technique to find the topics to which
    a document most suits to. 
    '''
    
    '''Build LDA model for K=5 topics '''
    lda_model = gensim.models.ldamodel.LdaModel(corpus=corpus,
                                               id2word=id2word,
                                               num_topics=5, 
                                               random_state=100, # similar to set.seed() so that results does not change 
                                               update_every=1, # update corpus after each pass
                                               chunksize=8, # earlier chunksize = 7 and num_topics = 5
                                               passes=100,  # 100 passes taking 8 samples(chunksize) in each iteration
                                               alpha='auto',
                                               per_word_topics=False)
    
    print("\nLDA Model topics output from Corpus[Latent Dirichlet Allocation Method] for:" + article_name + "\n")
    display(lda_model.print_topics())
    
    ''' Beta and Gamma Factor matrices - tokens to topics mapping, document to topics mapping respectively'''
    ## obtain the factor matrices - beta
    def build_beta_df(lda_model=lda_model, id2word=id2word):
        beta = lda_model.get_topics()  # shape (num of topics, vocabulary size).
        beta_df = pd.DataFrame(data=beta)
    
        # convert colnames in beta_df 2 tokens
        token2col = list(id2word.token2id)
        beta_df.columns = token2col
        rowNames=['topic' + format(x+1, '02d') for x in range(beta_df.shape[0])]
        rowNames_series = pd.Series(rowNames)
        beta_df.rename(index=rowNames_series, inplace=True)
        return(beta_df)
    
    #document to topics mapping respectively
    def build_gamma_df(lda_model, corpus0):
        gamma_doc = []  
        num_topics = lda_model.get_topics().shape[0]
        
        for doc in range(len(corpus0)):
            doc1 = corpus0[doc].split()
            bow_doc = id2word.doc2bow(doc1)
            gamma_doc0 = [0]*num_topics  
            gamma_doc1 = lda_model.get_document_topics(bow_doc)
            gamma_doc2_x = [x for (x,y) in gamma_doc1]
            gamma_doc2_y = [y for (x,y) in gamma_doc1]
            for i in range(len(gamma_doc1)):
                x = gamma_doc2_x[i]
                y = gamma_doc2_y[i]
                gamma_doc0[x] = y  
            gamma_doc.append(gamma_doc0)
            
        gamma_df = pd.DataFrame(data=gamma_doc)  # shape=num_docs x num_topics
        topicNames=['topic' + format(x+1, '02d') for x in range(num_topics)]
        topicNames_series = pd.Series(topicNames)
        gamma_df.rename(columns=topicNames_series, inplace=True)
        return(gamma_df)
    
    
    # Apply build_gamma_df function to get gamma factor matrices.
    gamma_df = build_gamma_df(lda_model=lda_model, corpus0=reviews) 
    print("\nGamma Matrix for: " + article_name)
    time.sleep(2)
    display(gamma_df)
    #gamma_df.iloc[:36, :]     
    
    
    '''Fit Metrices ''' 
    '''Compute Perplexity'''
    #print('\nPerplexity: ', lda_model.log_perplexity(corpus))
    # Perplexity:  -7.375374126997931 
    #Lower the better.
    row0 = gamma_df.values.tolist()
    row=[]
    for i in range(len(row0)):
        row1 = list(enumerate(row0[i]))
        row1_y = [y for (x,y) in row1]
        max_propn = sorted(row1_y, reverse=True)[0]
        row2 = [(i, x+1, y) for (x, y) in row1 if y==max_propn]
        row.append(row2)
 
    occurence_count = (gamma_df != 0).astype(int).sum(axis=1)
    
    document_occ_cnt_in_topics = pd.DataFrame(reviews, occurence_count).reset_index()
    document_occ_cnt_in_topics['tokens'] = pd.Series(corpus11)
    document_occ_cnt_in_topics.columns = ["cnt_of_topics_doc_appeared", "documents", "tokens"]
    print("\nDocument Occurrence in Topics (A document in multiple topics) in: " + article_name)
    display(document_occ_cnt_in_topics)
    
    sent_topics_df = pd.DataFrame()
    for row1 in row:
        for (doc_num, topic_num, prop_topic) in row1:
            wp = lda_model.show_topic(topic_num-1)
            topic_keywords = ", ".join([word for word, prop in wp])
            sent_topics_df = sent_topics_df.append(pd.Series([int(doc_num), int(topic_num), 
                                                              round(prop_topic,4), 
                                                              topic_keywords]), 
                                                           ignore_index=True)
        
    sent_topics_df.columns = ['Doc_num', 'Dominant_Topic', 'Perc_Contribution', 'Topic_Keywords']  
    print("\nDocuments with dominant topic and its percentage contribution in: "+article_name)
    display(sent_topics_df)
    
    
    # Add original text to the end of the output
    contents = pd.Series(reviews)
    sent_topics_df = pd.concat([sent_topics_df, contents], axis=1)
    sent_topics_df.columns = ['Doc_num', 'Dominant_Topic', 'Perc_Contribution', 'Topic_Keywords', 'orig_contents']
    #sent_topics_df.to_excel("C:/.../<.xlsx file name>", sheet_name = "BeforeIncidence")  # doctest: +SKIP
    return document_occ_cnt_in_topics, sent_topics_df


loop = (Zomato_data_during_incidence.dropna(), Zomato_data_after_incidence.dropna(), 
        Zomato_data_before_incidence.dropna(), Swiggy_data_during_incidence.dropna(), 
        Swiggy_data_after_incidence.dropna(), Swiggy_data_before_incidence.dropna())


for i in range(0, len(loop)):
    j=("zoma_during_incdnt_topicmdl", "zoma_after_incdnt_topicmdl", "zoma_before_incdnt_topicmdl",
       "swgy_during_incdnt_topicmdl", "swgy_after_incdnt_topicmdl", "swgy_before_incdnt_topicmdl")
    #j = ["zoma_bigram_during_incdnt_topicmdl", "TopicModelling_After", "TopicModelling_Before", "TopicModelling_During", "TopicModelling_After", "TopicModelling_Before"]
    print("\n\n")
    display("Get Topic Modelling Results for-"+str(j[i]))
    results = gensimmod(loop[i], article_name=str(j[i]))
    dominant_documents = results[0]  # first result of the function gensimmod ->document_occ_cnt_in_topics 
    documents_to_topic_mapping = results[1]

    #if i<3:
    with pd.ExcelWriter("C:/.../Swiggy-Zomato-Food-Aggregators-Analysis/articles_data/lda_output/" + j[i] + ".xlsx") as writer:
        dominant_documents.to_excel(writer, sheet_name= 'dominant_documents')  
        documents_to_topic_mapping.to_excel(writer, sheet_name= 'documents_to_topics')
            
```
